Ans5 :Sampling is the process of selecting a subset (sample) from a larger group (population) to analyze and make inferences about the whole population.
Sampling is essential when studying an entire population is impractical or impossible due to time, cost, or logistical constraints.

Ans6 : Sampling methods are divided into random (probability) and non-random (non-probability) categories:

# Random Sampling Methods:

Simple Random Sampling: Every member has an equal chance of being selected.

Systematic Sampling: Every kth member is selected from a list.

Stratified Sampling: The population is divided into subgroups (strata) and sampled from each.

Cluster Sampling: The population is divided into clusters, some of which are randomly selected, and all members of those clusters are surveyed.

# Non-Random Sampling Methods:

Convenience Sampling: Samples are taken from a group that is easy to access.

Judgmental/Purposive Sampling: The researcher selects samples based on their knowledge/judgment.

Quota Sampling: Ensures certain traits are represented by setting quotas.

Snowball Sampling: Existing subjects recruit future subjects, often used in hidden populations.

Ans7 :Difference between random and non-random sampling :
# Random Sampling

Every individual has a known and equal chance of selection

Reduces bias and is more representative of the population

Used in scientific research and statistical inference

# Non-Random Sampling

Individuals are selected based on non-random criteria

More prone to bias and may not represent the population well.

Often used in exploratory research or when random sampling is not feasible.

Ans 18: Skewness measures the asymmetry of a data distribution.

If the distribution is symmetric, skewness is 0.

If it's not symmetric, it's either positively or negatively skewed.

Ans 19:
***Positive Skew*** (Right-skewed): Tail is longer on the right side, Most data is on the left e.g Income distribution


***Negative Skew*** (Left-skewed)	Tail is longer on the left side, Most data is on the right, e.g Age at retirement


Ans 20: Kurtosis measures the "tailedness" of a distributionâ€”how heavy or light the tails are compared to a normal distribution.

Mesokurtic (kurtosis â‰ˆ 3): Normal distribution.

Leptokurtic (kurtosis > 3): Heavier tails (more outliers).

Platykurtic (kurtosis < 3): Lighter tails (fewer outliers).

It helps assess the risk of extreme values.



Ans 21: Covariance measures how two variables change together.

Positive covariance: Both variables tend to increase together.

Negative covariance: One increases while the other decreases.

It's used to understand the direction of the relationship between variables.

Ans 22: Correlation measures the strength and direction of a linear relationship between two variables.

Ranges from -1 to +1:

+1: Perfect positive correlation

0: No correlation

-1: Perfect negative correlation

Common measure: Pearson correlation coefficient (r).

Ans23: difference between covariance and correlation:-

# Covariance

Measures direction of relationship

Value depends on the units of variables

Hard to interpret magnitude

Can be any number

# Correlation

Measures direction and strength

Standardized, unitless

Easy to interpret (between -1 and 1)

Always between -1 and 1

```python
#Ans8
import numpy as np

# Positive skew (right-skewed)
pos_skewed = np.random.exponential(scale=2, size=1000)

# Negative skew (left-skewed)
neg_skewed = -np.random.exponential(scale=2, size=1000)

# Plotting
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
sns.histplot(pos_skewed, kde=True, ax=axes[0], color='orange')
axes[0].set_title("Positive Skew")
sns.histplot(neg_skewed, kde=True, ax=axes[1], color='blue')
axes[1].set_title("Negative Skew")
plt.show()
```

```python
#Ans12
import pandas as pd
import numpy as np
import random

# Create sample dataset
data = pd.DataFrame({'ID': range(1, 21)})

# Simple Random Sampling (5 samples)
simple_random_sample = data.sample(n=5, random_state=1)

# Systematic Sampling (every 4th element)
k = 4
start = random.randint(0, k-1)
systematic_sample = data.iloc[start::k].head(5)

print("Simple Random Sample:\n", simple_random_sample)
print("\nSystematic Sample:\n", systematic_sample)
```

```python
#Ans13
import statistics

# Midpoints and frequencies
midpoints = [5, 15, 25, 35, 45]
frequencies = [2, 5, 8, 4, 1]

# Mean for grouped data
total_freq = sum(frequencies)
mean = sum([midpoints[i] * frequencies[i] for i in range(len(midpoints))]) / total_freq

# Cumulative frequencies for median and mode can be handled more easily with pandas or custom logic
print("Grouped Mean:", mean)
```

```python
#Ans14
import numpy as np
import statistics

# Simulate normal distribution data
data = np.random.normal(loc=50, scale=10, size=1000)

# Central tendency
mean = np.mean(data)
median = np.median(data)
mode = statistics.mode(data)

# Dispersion
variance = np.var(data)
std_dev = np.std(data)
range_val = np.max(data) - np.min(data)

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
print("Variance:", variance)
print("Standard Deviation:", std_dev)
print("Range:", range_val)
```

```python
#Ans20
import seaborn as sns
import matplotlib.pyplot as plt

# Simulated dataset
df = pd.DataFrame({
    'Math': np.random.randint(50, 100, 100),
    'Science': np.random.randint(40, 100, 100),
    'English': np.random.randint(60, 100, 100)
})

# Covariance matrix
cov_matrix = df.cov()

# Heatmap
sns.heatmap(cov_matrix, annot=True, cmap='coolwarm')
plt.title("Covariance Heatmap")
plt.show()
```

```python
#Ans22
import numpy as np

# Generate dataset
data = np.random.randint(10, 100, size=20)

# Compute variance and standard deviation
variance = np.var(data, ddof=1)  # sample variance
std_dev = np.std(data, ddof=1)   # sample standard deviation

print("Dataset:", data)
print("Variance:", variance)
print("Standard Deviation:", std_dev)
```

```python
#Ans24
from scipy.stats import pearsonr, spearmanr

# Generate two correlated datasets
x = np.random.randint(10, 100, size=30)
y = x + np.random.normal(0, 10, size=30)

# Pearson Correlation
pearson_corr, _ = pearsonr(x, y)

# Spearman Correlation
spearman_corr, _ = spearmanr(x, y)

print("Pearson Correlation Coefficient:", pearson_corr)
print("Spearman Correlation Coefficient:", spearman_corr)
```
